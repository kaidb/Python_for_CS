```python
import time
```

```python
def oracle(inp):
    return inp == 'zbcde'
```

```python
alphabet = [chr(97 + i)  for i in range(26)]


class BruteForce:
    def __init__(self, alphabet, size):
        self.alphabet = alphabet
        self.MAX = len(alphabet)
        self.size = size
        self.state = [0 for i in range(size)]
        self.Going = True
        total = size ** len(alphabet)
        print("Initilized Brute Force generator.")
        print("Total Number of querries worst case: {}".format(total))
        
    def caryOver(self, index):
        # Cary the 1
        if self.state[index] == self.MAX:
            self.state[index] = 0
            if index ==-1 * self.size:
                # The list has finished 
                return False
            index -=1
            self.state[index] +=1
            return self.caryOver(index)
        return True
    
    def stillGoing(self):
        return self.Going
    
    def nextGuess(self):
        guess = ''.join([self.alphabet[i] for i in self.state])
        self.state[-1]+=1
        self.going = self.caryOver(-1)
        return guess
        

```

```python
# Brute force example:
start = time.time()
BF = BruteForce(alphabet, 5)
while BF.stillGoing:
    guess = BF.nextGuess()
    #print(guess)
    if oracle(guess):
        print("Found password '{}' in {} seconds".format(guess, time.time()-start))
        break
        
    
```

```python
import threading
from queue import Queue
```

```python
BF = BruteForce(alphabet, 5)
bf_lock = threading.Lock()
print_lock = threading.Lock()
start = 0

def threader():
    while BF.Going:
        with bf_lock:
            guess = BF.nextGuess()
        if oracle(guess):
            BF.Going = False
            print("Password is {} in {} seconds".format(guess, time.time() - start))

```

```python
n_threads = 25
start = time.time()
for x in range(n_threads):
    t = threading.Thread(target=threader)

    # classifying as a daemon, so they will die when the main dies
    t.daemon = True

    # begins, must come after daemon definition
    t.start()
print(time.time()-start)
# Notice that is can actually be slower if you're not careful!
# This is because it occisionally uses virtual threads!
```
